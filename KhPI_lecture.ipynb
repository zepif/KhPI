```python
import numpy as np
import matplotlib.pyplot as plt

x1 = 0
x2 = 4

x_interval1 = np.linspace(-15, x1 - 0.1, 400)  # Для x <= 0
x_interval2 = np.linspace(x1 + 0.1, x2 - 0.1, 400)  # Для 0 < x < 4
x_interval3 = np.linspace(x2 + 0.1, 6, 400)  # Для x > 4

def f(x):
    return np.piecewise(x, [x <= x1, (x1 < x) & (x < x2), x >= x2], [lambda x: np.cos(x), lambda x: 2 / (x + 1), lambda x: x**2 - 3])

y_interval1 = f(x_interval1)
y_interval2 = f(x_interval2)
y_interval3 = f(x_interval3)

plt.figure(figsize=(10, 6))
plt.plot(x_interval1, y_interval1, label='f(x) = cos(x), x <= 0', linestyle='-', color='blue')
plt.plot(x_interval2, y_interval2, label='f(x) = 2/(x + 1), 0 < x < 4', linestyle='-', color='green')
plt.plot(x_interval3, y_interval3, label='f(x) = x^2 - 3, x > 4', linestyle='-', color='red')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='best')

plt.xlim(-15, 6)
plt.ylim(-2, 20)

plt.axvline(x=x1, color='gray', linestyle='--', label='x = 0')
plt.axvline(x=x2, color='gray', linestyle='--', label='x = 4')

plt.title('График функции f(x)')
plt.grid(True)
plt.show()
```

```python
import sympy as sp

x1, x2, x3 = sp.symbols('x1 x2 x3')

eq1 = sp.Eq(3*x1 + 2*x2 + x3, 7)
eq2 = sp.Eq(x2 + x3, 8)
eq3 = sp.Eq(x1 + x2, 2)

solution = sp.solve((eq1, eq2, eq3), (x1, x2, x3))

print("Решение системы уравнений:")
print(solution)

n = sp.symbols('n')

eq1 = sp.Eq(3*x1 + 2*x2 + x3, n)
eq2 = sp.Eq(x2 + x3, 8)
eq3 = sp.Eq(x1 + x2, 2)

solution = sp.solve((eq1, eq2, eq3), (x1, x2, x3))

print("Решение системы уравнений:")
print(solution)
```

```python
import scipy.optimize as opt
import math

def equation_to_solve(x):
    return math.sin(1 / x) - 0.5

solution = opt.root_scalar(equation_to_solve, bracket=[0.01, 1.0])

if solution.converged:
    x = solution.root
    print(f"Solution found using SciPy: x = {x}")
else:
    print("No solution found within the specified bracket.")
```

```python
import turtle

def apply_rules(ch):
    if ch == 'F':
        return 'F-F++F-F'
    else:
        return ch

def generate_l_system(axiom, iterations):
    current_string = axiom
    for _ in range(iterations):
        next_string = ''.join(apply_rules(ch) for ch in current_string)
        current_string = next_string
    return current_string

def draw_l_system(turtle, instructions, angle, length):
    for instruction in instructions:
        if instruction == 'F':
            turtle.forward(length)
        elif instruction == '+':
            turtle.left(angle)
        elif instruction == '-':
            turtle.right(angle)

def draw_christmas_tree(iterations, length, angle):
    axiom = 'F'
    instructions = generate_l_system(axiom, iterations)

    turtle.speed(2)
    turtle.penup()
    turtle.goto(0, -200)
    turtle.pendown()

    draw_l_system(turtle, instructions, angle, length)

    turtle.hideturtle()
    turtle.done()

if __name__ == "__main__":
    iterations = 4
    length = 10
    angle = 60

    draw_christmas_tree(iterations, length, angle)
```

```python
from sympy import symbols, integrate, ln, pprint

x = symbols('x')
expression = x**2 * ln(x + 1)

result = integrate(expression, x)
print("Первообразная x^2 * ln(x + 1):")
pprint(result)

a, b = 0, 0.9
definite_result = integrate(expression, (x, a, b))
print(f"Интеграл x^2 * ln(x + 1) для a={a} и b={b}: {definite_result}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, ln

x = symbols('x')

expression = x**2 * ln(x + 1)

derivative_first_order = diff(expression, x)
derivative_second_order = diff(derivative_first_order, x)

f = lambda x_val: expression.subs(x, x_val)
f_prime = lambda x_val: derivative_first_order.subs(x, x_val)
f_double_prime = lambda x_val: derivative_second_order.subs(x, x_val)

x_vals = np.linspace(0.01, 0.9, 100)

y_vals = np.vectorize(f)(x_vals)
y_prime_vals = np.vectorize(f_prime)(x_vals)
y_double_prime_vals = np.vectorize(f_double_prime)(x_vals)

plt.figure(figsize=(10, 6))

plt.plot(x_vals, y_vals, label=r'$x^2 \ln(x+1)$')
plt.plot(x_vals, y_prime_vals, label=r'$\frac{d}{dx}(x^2 \ln(x+1))$', linestyle='--', color='orange')
plt.plot(x_vals, y_double_prime_vals, label=r'$\frac{d^2}{dx^2}(x^2 \ln(x+1))$', linestyle='-.', color='green')

plt.title('Графики функции и ее производных')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
